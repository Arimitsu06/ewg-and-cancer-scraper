```python
import requests
from bs4 import BeautifulSoup
import time
import math
import re
import tqdm
from tqdm import tqdm
import pandas as pd
```

```python
# for URL in urls:
URL = "https://statecancerprofiles.cancer.gov/incidencerates/index.php?stateFIPS=36&areatype=county&cancer=001&race=00&sex=0&age=001&stage=999&year=0&type=incd&sortVariableName=name&sortOrder=asc#results"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
rows = soup.find_all("tr")
```

```python
rows
```

```python
str(rows[3]).split("<strong>")[1].replace("\n", "").strip(" ").split("  ")[0]
```

```python
cancer_rates_dict = {"county": [], "cases_per_100k": [], "average_annual_count_5y": []}
```

```python
for i in range(3, len(rows)):
    county_name = str(rows[i]).split("<strong>")[1].replace("\n", "").strip(" ").split("  ")[0]
    cases_per_100k = float(str(rows[i]).split("<strong>")[3].replace("\t", "").replace(" ", "").replace("\n", "").split("(")[0])
    avg_ann_count = int(str(rows[i]).split("<strong>")[5].replace("\t", "").replace("\n", "").replace(" ", "").split("<")[0].replace(",", ""))
    cancer_rates_dict['cases_per_100k'].append(cases_per_100k)
    cancer_rates_dict['county'].append(county_name)
    cancer_rates_dict['average_annual_count_5y'].append(avg_ann_count)
```

```python
cancer_rates_dict
```

```python
cancer_df = pd.DataFrame.from_dict(cancer_rates_dict)
```

```python
cancer_df.to_csv("ny_cancer_rates.csv", index=False)
```
